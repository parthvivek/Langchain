## Prompt Template 
### Prompt templating is the practice of creating parameterized prompts where: 
#### The structure stays fixed
#### The inputs change dynamically
#### Instead of writing prompts manually, you define a template + variables.

```python
from langchain.prompts import PromptTemplate
prompt_template_name = PromptTemplate(
    input_variables = ["country"],
    template = "Can you tell the capital of {country}?"
)
prompt1 = prompt_template_name.format(country = "india")

def gemini(prompt):
    llm = ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        google_api_key=api_key,
        temperature=0
    )
     # this is the prompt variable
    messages = [
        SystemMessage(content="You are a helpful AI assistant who answers short and crisp"),
        HumanMessage(content=prompt)
    ]

    # Test it
    response = llm.invoke(messages)
    return response.content

gemini(prompt1)
```
